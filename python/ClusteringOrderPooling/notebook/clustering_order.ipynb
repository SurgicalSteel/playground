```python
import pandas as pd
from sklearn.cluster import KMeans
import folium

file_url = "../dataset/orders.csv"
data = pd.read_csv(file_url)
features = data[['Latitude', 'Longitude']]
print(features)

kmeans = KMeans(
    init="random",
    n_clusters=20,
    n_init=10,
    max_iter=1000,
    random_state=42
)

# do clustering
kmeans.fit(features)
# save results
labels = kmeans.labels_

# send back into dataframe and display it
data['Cluster'] = labels
# display the number of mamber each clustering
_clusters = data.groupby('Cluster')['OrderId'].count()
print(_clusters)

colors = ['red', 'blue', 'green', 'purple', 'orange', 'darkred', \
     'lightred', 'beige', 'darkblue', 'darkgreen', 'cadetblue', \
     'darkpurple', 'pink', 'lightblue', 'lightgreen', 'gray', \
     'black', 'lightgray', 'red', 'blue', 'green', 'purple', \
     'orange', 'darkred', 'lightred', 'beige', 'darkblue', \
     'darkgreen', 'cadetblue', 'darkpurple','pink', 'lightblue', \
     'lightgreen', 'gray', 'black', 'lightgray' ]
lat = -7.802942 #data.iloc[0]['Latitude']
lng =  110.364549 # data.iloc[0]['Longitude']
map = folium.Map(location=[lng, lat], zoom_start=9)
for _, row in data.iterrows():
    folium.CircleMarker(
        location=[row["Latitude"], row["Longitude"]],
        radius=12, 
        weight=2, 
        fill=True, 
        fill_color=colors[int(row["Cluster"])],
        color=colors[int(row["Cluster"])]
    ).add_to(map)
    
map

```
